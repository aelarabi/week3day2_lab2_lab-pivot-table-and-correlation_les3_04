#### We want to know the information about the mean salary and number of employees in each department and for each job title.

```python
# your answer here
data.pivot_table(index= ['Department','Title'], values = ['Salary','Name'], aggfunc = {'Salary':np.mean, 'Name':'count'}).rename(columns={'Name':"number_of_employees"})
```

#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python
# your answer here
data.pivot_table(index= ['Department','Title'], values = ['Salary'], aggfunc = {'Salary': [np.median, min, max, np.mean,np.std]})

#There is no standard deviation for a single observation. There is only one Hr VP and 1 IT VP. if we drop the department
# from the pivot, we would get a std of 0 for the reported salaries of both VPs. 
```

#### Based on your comments, fill the missing values with an appropriate value.

```python
# your answer here
# The logical fill is 0 given the lack of variance in VP mean. 
#import math
data_pivot=data.pivot_table(index= ['Department','Title'], values = ['Salary'], aggfunc = {'Salary': [np.median, min, max, np.mean,np.std]}, fill_value=0)
round(data_pivot,2)
```

#### The stake holders want to know for each department the number of employees and how much money is spend on salaries. Could you provide that information? 

```python
# your answer here
data.pivot_table(index= ['Department'], values = ['Salary','Name'], aggfunc = {'Salary': sum, 'Name':'count'}).rename({'Name':"number_of_employees"},axis=1)
```

#### For each Department and Title, we want to know the the total years of work experience and the mean salary.

```python
# your answer here
data.pivot_table(index= ['Department','Title'], values = ['Years','Salary'], aggfunc = {'Years':sum,'Salary':np.mean}).rename(columns={'Years':"Total years of work experience"})
```
